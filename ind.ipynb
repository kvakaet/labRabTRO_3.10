```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Выделение фрагмента фотографии

```python
img = cv2.imread('images/kreml.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

image = cv2.rectangle(img, (420, 460), (370, 390), (0, 0, 255), 2)
plt.axis('off')
plt.title('Выделенный номер')
plt.imshow(image);
```

# Взятие контура выделенного знака
Сначала возьмём сам фрагмент

```python
crop = img[420:460, 370:390]
piece = cv2.resize(
    crop, 
    (200, 100), 
    interpolation=cv2.INTER_LINEAR
)

plt.axis('off')
plt.title('Выделенный номер')
plt.imshow(piece);
```

Теперь возьмём контур

```python
img = cv2.cvtColor(piece, cv2.COLOR_RGB2GRAY)

ret, thresh = cv2.threshold(img, 127, 255, 0)
contours, hierarchy = cv2.findContours(thresh, 5, 5)
cv2.drawContours(piece, contours, -1, (0,255,0), 1)

plt.axis("off")
plt.title('Контур выделенного номера')
plt.imshow(piece);
```
